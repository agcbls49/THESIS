```python
# Create the YAML file needed for YOLO
import yaml

data = {
    'path': '/content/dataset/dataset',
    'train': 'images/train',
    'val': 'images/val',
    'test': 'images/test',
    'names': {
        0: 'backpack',
        1: 'book',
        2: 'cup',
        3: 'glasses',
        4: 'headphone',
        5: 'key',
        6: 'phone',
        7: 'remote',
        8: 'wallet',
        9: 'water bottle'
    }
}

with open('/content/data.yaml', 'w') as f:
    yaml.dump(data, f, default_flow_style=False)

print("data.yaml created!")
```

## **YOLO v5 NANO MODEL**

```python
# Train YOLOv5 Nano
# imgsz=640 means images are resized to be 640x640 and batch=16 means it will look at 16 images at a time before updating what it learned
# name=yolov5n_run is the name of the folder where the results will be saved
!yolo detect train data=/content/data.yaml model=yolov5n.pt epochs=50 imgsz=640 batch=16 name=yolov5n_run
```

## **INTERPRETATION OF RESULTS - YOLO V5 NANO (BASED ON THE CSV FILE)**
Precision = 0.98137 <br>
Recall = 0.92967 <br>
mAP50 = 0.97396 <br>
mAP50-95 = 0.85258 <br>

---

When the model draws a box, 98.1% of the boxes drawn are correctly matched from the annotated boxes or the ground truth. Out of all the annotated ground truth, 92.9% of them were detected by the model. When the model draws a box and it has at least a 50% overlap with the ground truth box, the average precision across all classes is 97.4%. Lastly, when the model draws a box and the overlap requirement goes from 50% to 95% in 5% steps, the average precision across all classes averaged across all those steps is 85.3%.

---

The model performs very well and it rarely draws wrong boxes (only 1.9% false alarms). It finds most of the ground truth boxes (misses only 7.1%).
