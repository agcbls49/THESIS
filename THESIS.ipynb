### Creating a YAML file
This step creates a YAML file which is needed for training YOLO models. It contains the path of the unzipped dataset, where the train, validation and test images are located, and the names of the classes along with their class numbers.

```python
import yaml
import os

data = {
    'path': '/content/dataset/dataset',
    'train': 'images/train',
    'val': 'images/val',
    'test': 'images/test',
    'nc': 20,
    'names': [
        'backpack',
        'book',
        'cup',
        'glasses',
        'headphone',
        'key',
        'phone',
        'remote',
        'wallet',
        'water bottle',
        'body spray',
        'card',
        'charger',
        'comb',
        'flashlight',
        'glasses case',
        'medicine',
        'nail clipper',
        'shoe',
        'watch'
    ]
}

yaml_path = '/content/data.yaml'

# write file
with open(yaml_path, 'w') as f:
    yaml.dump(data, f)

print("written")

# verify file exists
print("exists:", os.path.exists(yaml_path))

# verify size
print("size:", os.path.getsize(yaml_path))

# print contents
with open(yaml_path, 'r') as f:
    content = f.read()

print("CONTENT:")
print(content)
```

## **YOLO v5 NANO MODEL**

```python
!yolo detect train data=/content/data.yaml model=yolov5nu.pt epochs=50 imgsz=640 batch=16 name=yolov5n_run
```

### **YOLO v5 NANO RESULTS**
Precision = 0.98137 <br>
Recall = 0.92967 <br>
mAP50 = 0.97396 <br>
mAP50-95 = 0.85258 <br>

---
### **INTERPRETATION**

When the model draws a box, 98.1% of the boxes drawn are correctly matched from the annotated boxes or the ground truth. Out of all the annotated ground truth, 92.9% of them were detected by the model. When the model draws a box and it has at least a 50% overlap with the ground truth box, the average precision across all classes is 97.4%. Lastly, when the model draws a box and the overlap requirement goes from 50% to 95% in 5% steps, the average precision across all classes averaged across all those steps is 85.3%.

---
### **OVERALL**
The model performs very well and it rarely draws wrong boxes (only 1.9% false alarms). It finds most of the ground truth boxes (misses only 7.1%).

---
## **TABLE INTERPRETATION**

All four models demonstrate strong performance, but they differ in how they balance precision, recall, and overall detection quality. **YOLOv8n** achieves the highest precision (98.6%) and the lowest false alarm rate (1.4%), making it the most reliable in avoiding incorrect detections, although it has slightly lower recall (92.9%) and misses the same proportion of objects as YOLOv5n. **YOLOv5n** provides a very balanced performance, with high precision (98.1%), solid recall (93%), and competitive mAP scores, making it a stable and consistent option overall. **YOLOv11n** stands out by achieving the highest mAP50 (97.7%) and mAP50–95 (86.6%), indicating the best overall detection quality and bounding box accuracy; it also improves recall (93.7%) compared to v5n and v8n, resulting in fewer missed objects, though at the cost of slightly higher false alarms (2.9%). In contrast, **YOLOv26n** prioritizes recall (94.4%), achieving the fewest misses (5.6%), but this comes with a significant drop in precision (92.7%) and the highest false alarm rate (7.3%), as well as the lowest mAP scores, indicating weaker overall accuracy.



---

# **CONCLUSION**

Considering all metrics, **YOLOv11n is the best overall model**, as it provides the highest detection accuracy (mAP), a strong balance between precision and recall, and improved object coverage with only a minimal increase in false positives.
