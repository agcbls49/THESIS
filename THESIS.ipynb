### Creating a YAML file
This step creates a YAML file which is needed for training YOLO models. It contains the path of the unzipped dataset, where the train, validation and test images are located, and the names of the classes along with their class numbers.

```python
import yaml
import os

data = {
    'path': '/content/dataset/dataset',
    'train': 'images/train',
    'val': 'images/val',
    'test': 'images/test',
    'nc': 20,
    'names': [
        'backpack',
        'book',
        'cup',
        'glasses',
        'headphone',
        'key',
        'phone',
        'remote',
        'wallet',
        'water bottle',
        'body spray',
        'card',
        'charger',
        'comb',
        'flashlight',
        'glasses case',
        'medicine',
        'nail clipper',
        'shoe',
        'watch'
    ]
}

yaml_path = '/content/data.yaml'

# write file
with open(yaml_path, 'w') as f:
    yaml.dump(data, f)

print("written")

# verify file exists
print("exists:", os.path.exists(yaml_path))

# verify size
print("size:", os.path.getsize(yaml_path))

# print contents
with open(yaml_path, 'r') as f:
    content = f.read()

print("CONTENT:")
print(content)
```

# **COMPARISON TABLE**


| Model | Precision | Recall | mAP50 | mAP50-95 | False Alarm Rate | Miss Rate |
|-------|-----------|--------|-------|----------|-----------------|-----------|
| v5n | 86.92% | 86.89% | 91.78% | 77.15% | 13.08% | 13.11% |
| v8n | 86.98% | 90.89% | 92.57% | 79.5% | 13.02% | 9.11% |
| v11n (50 Epochs) | 90.87% | 89.31% | 93.88% | 80.93% | 9.13% | 10.69% |
| v26n | 90.76% | 83.61% | 89.76% | 77.36% | 9.24% | 16.39% |

---
## **TABLE INTERPRETATION**

**v5n** obtained a Precision of 86.92% and a Recall of 86.89%, indicating that the model was able to correctly identify most objects while maintaining a balanced detection capability. The model achieved an mAP50 of 91.78% and an mAP50-95 of 77.15%, showing acceptable detection accuracy across different IoU thresholds. However, the model recorded a False Alarm Rate of 13.08% and a Miss Rate of 13.11%, which are both higher than those of v11n (50 Epochs) and v8n in terms of error performance. This suggests that v5n produced more incorrect detections and missed more objects compared to some of the newer models.
<br><br>

**v8n** achieved a Precision of 86.98% and the highest Recall among all models at 90.89%, indicating that it was the most effective model in detecting actual objects with the fewest missed detections. The model also recorded an mAP50 of 92.57% and an mAP50-95 of 79.5%, both of which are higher than those of v5n and v26n. In addition, v8n obtained the lowest Miss Rate at 9.11%, confirming its strong detection capability. However, its False Alarm Rate of 13.02% remained higher than those of v11n (50 Epochs) and v26n, meaning that although the model detected more objects, it also produced more false detections compared to those models.
<br><br>

**v11n (50 Epochs)** demonstrated the highest Precision at 90.87%, indicating that it produced the most accurate positive detections among all models. It also achieved the highest mAP50 at 93.88% and the highest mAP50-95 at 80.93%, reflecting the strongest overall object detection performance and localization accuracy across different IoU thresholds. Furthermore, the model recorded the lowest False Alarm Rate at 9.13%, showing that it generated the fewest incorrect detections. Although its Recall of 89.31% and Miss Rate of 10.69% were not the best overall, the results still indicate strong and balanced performance across all evaluation metrics.
<br><br>

**v26n** obtained a Precision of 90.76%, which is slightly lower than v11n (50 Epochs) but higher than both v5n and v8n. Despite this, the model recorded the lowest Recall at 83.61%, indicating that it missed more actual objects than the other models. This is further supported by its Miss Rate of 16.39%, which is the highest among all models. The model achieved an mAP50 of 89.76% and an mAP50-95 of 77.36%, both of which are lower than those of v5n, v8n, and v11n (50 Epochs). However, v26n maintained a relatively low False Alarm Rate of 9.24%, second only to v11n (50 Epochs), suggesting that while the model was more conservative in making detections, it produced fewer false positives.
