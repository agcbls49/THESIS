### Creating a YAML file
This step creates a YAML file which is needed for training YOLO models. It contains the path of the unzipped dataset, where the train, validation and test images are located, and the names of the classes along with their class numbers.

```python
import yaml
import os

data = {
    'path': '/content/dataset/dataset',
    'train': 'images/train',
    'val': 'images/val',
    'test': 'images/test',
    'nc': 20,
    'names': [
        'backpack',
        'book',
        'cup',
        'glasses',
        'headphone',
        'key',
        'phone',
        'remote',
        'wallet',
        'water bottle',
        'body spray',
        'card',
        'charger',
        'comb',
        'flashlight',
        'glasses case',
        'medicine',
        'nail clipper',
        'shoe',
        'watch'
    ]
}

yaml_path = '/content/data.yaml'

# write file
with open(yaml_path, 'w') as f:
    yaml.dump(data, f)

print("written")

# verify file exists
print("exists:", os.path.exists(yaml_path))

# verify size
print("size:", os.path.getsize(yaml_path))

# print contents
with open(yaml_path, 'r') as f:
    content = f.read()

print("CONTENT:")
print(content)
```

# **RUNNING THE MODELS**

### **YOLO v5 NANO MODEL**

```python
!yolo detect train data=/content/data.yaml model=yolov5nu.pt epochs=50 imgsz=640 batch=16 name=yolov5n_run
```

#### **YOLO v5 NANO RESULTS**
Precision = 0.9094 <br>
Recall = 0.82526 <br>
mAP50 = 0.90511 <br>
mAP50-95 = 0.75036 <br>

---
##### **INTERPRETATION**

When the model draws a box, 90.94% of the boxes drawn are correctly matched from the annotated boxes or the ground truth. Out of all the annotated ground truth, 82.53% of them were detected by the model. When the model draws a box and it has at least a 50% overlap with the ground truth box, the average precision across all classes is 90.51%. Lastly, when the model draws a box and the overlap requirement goes from 50% to 95% in 5% steps, the average precision across all classes averaged across all those steps is 75.04%.

---
##### **OVERALL**
The model performs very well and it rarely draws wrong boxes (only 9.06% false alarms). It finds most of the ground truth boxes (misses only 17.47%).

# **COMPARISON TABLE**


| Model | Precision | Recall | mAP50 | mAP50-95 | False Alarm Rate | Miss Rate |
|-------|-----------|--------|-------|----------|-----------------|-----------|
| v5n | 90.94% | 82.53% | 90.51% | 75.04% | 9.06% | 17.47% |
| v8n | 90.56% | 89.70% | 92.95% | 78.36% | 9.44% | 10.30% |
| v11n (50 Epochs) | 91.33% | 88.25% | 92.50% | 79.07% | 8.67% | 11.75% |
| v26n | 87.01% | 79.83% | 87.79% | 74.51% | 12.99% | 20.17% |

---
## **TABLE INTERPRETATION**

All four models demonstrate strong performance, but they differ in how they balance precision, recall, and overall detection quality. **YOLOv11n (50 Epochs)** achieves the highest precision (91.33%) and the lowest false alarm rate (8.67%), making it the most reliable in avoiding incorrect detections, although it has a slightly lower recall (88.25%) compared to YOLOv8n.
<br><br>

**YOLOv5n** delivers moderate performance, with a precision of 90.94% and a false alarm rate of 9.06%, but records the second highest miss rate (17.47%) and lower mAP scores (mAP50: 90.51%, mAP50-95: 75.04%), indicating weaker recall performance compared to v8n and v11n.
<br><br>

**YOLOv8n** provides a very balanced performance, with the highest recall (89.70%) and competitive mAP scores (mAP50: 92.95%, mAP50-95: 78.36%), making it a stable and consistent option overall, despite a slightly lower precision (90.56%).
<br><br>

In contrast, **YOLOv26n** has the lowest recall (79.83%), resulting in the highest miss rate (20.17%), along with the lowest precision (87.01%), the highest false alarm rate (12.99%), and the lowest mAP scores (mAP50: 87.79%, mAP50-95: 74.51%), indicating the weakest overall detection accuracy among the four models.

# **CONCLUSION**

Considering all metrics, **YOLOv11n is the best overall model**, as it provides the highest detection accuracy (mAP), a strong balance between precision and recall, and improved object coverage with only a minimal increase in false positives.
