### Creating a YAML file
This step creates a YAML file which is needed for training YOLO models. It contains the path of the unzipped dataset, where the train, validation and test images are located, and the names of the classes along with their class numbers.

```python
import yaml
import os

data = {
    'path': '/content/dataset/dataset2',
    'train': 'images/train',
    'val': 'images/val',
    'test': 'images/test',
    'nc': 20,
    'names': [
        'backpack',
        'book',
        'cup',
        'glasses',
        'headphone',
        'key',
        'phone',
        'remote',
        'wallet',
        'water bottle',
        'body spray',
        'card',
        'charger',
        'comb',
        'flashlight',
        'glasses case',
        'medicine',
        'nail clipper',
        'shoe',
        'watch'
    ]
}

yaml_path = '/content/data.yaml'

# write file
with open(yaml_path, 'w') as f:
    yaml.dump(data, f)

print("written")

# verify file exists
print("exists:", os.path.exists(yaml_path))

# verify size
print("size:", os.path.getsize(yaml_path))

# print contents
with open(yaml_path, 'r') as f:
    content = f.read()

print("CONTENT:")
print(content)
```

# **CONCLUSION**
The refined **150-epoch YOLOv11n** model was selected for deployment in the Android application. Although the 100-epoch model achieved a slightly higher precision (94.77% vs. 94.56%) and a slightly lower false alarm rate (5.23% vs. 5.44%), the differences were minimal at only 0.21. In contrast, the 150-epoch model demonstrated higher recall (89.41% vs. 87.17%), higher mAP@50 (93.06% vs. 92.96%), higher mAP@50–95 (79.44% vs. 79.31%), and a lower miss rate (10.59% vs. 12.83%).

<br>

Since **mAP@50–95** is the primary evaluation metric for model selection, the 150-epoch model achieved the best overall object detection performance. Additionally, its higher recall and lower miss rate indicate that it detects more target objects while missing fewer instances, which is particularly important for an object-finding application designed to assist visually impaired users. Therefore, despite the negligible decrease in precision and slight increase in the false alarm rate, the overall improvements in detection performance justify selecting the refined 150-epoch model for deployment.
