### Creating a YAML file
This step creates a YAML file which is needed for training YOLO models. It contains the path of the unzipped dataset, where the train, validation and test images are located, and the names of the classes along with their class numbers.

```python
import yaml
import os

data = {
    'path': '/content/dataset/dataset1',
    'train': 'images/train',
    'val': 'images/val',
    'test': 'images/test',
    'nc': 20,
    'names': [
        'backpack',
        'book',
        'cup',
        'glasses',
        'headphone',
        'key',
        'phone',
        'remote',
        'wallet',
        'water bottle',
        'body spray',
        'card',
        'charger',
        'comb',
        'flashlight',
        'glasses case',
        'medicine',
        'nail clipper',
        'shoe',
        'watch'
    ]
}

yaml_path = '/content/data.yaml'

# write file
with open(yaml_path, 'w') as f:
    yaml.dump(data, f)

print("written")

# verify file exists
print("exists:", os.path.exists(yaml_path))

# verify size
print("size:", os.path.getsize(yaml_path))

# print contents
with open(yaml_path, 'r') as f:
    content = f.read()

print("CONTENT:")
print(content)
```

## **TABLE INTERPRETATION**

**YOLO v5n** <br>
v5n achieved 93.64% Precision, 85.33% Recall, 91.49% mAP@50, and 75.67% mAP@50–95, with a 6.36% False Alarm Rate and a 14.67% Miss Rate. These results show that v5n missed more objects than the other models and had lower detection performance at stricter IoU thresholds.
<br>

**YOLO v8n** <br>
v8n increased Recall to 87.62%, mAP@50 to 92.48%, and mAP@50–95 to 77.79% while maintaining Precision at 93.65%. It also recorded the lowest False Alarm Rate (6.35%) and reduced the Miss Rate to 12.28%. These results show that v8n detected more objects than v5n while recording a lower False Alarm Rate than v5n and v26n.
<br>

**YOLO v11n** <br>
v11n recorded 93.30% Precision, 88.92% Recall, 92.75% mAP@50, and 78.20% mAP@50–95. Its False Alarm Rate was 6.70%, and its Miss Rate was 11.08%, the lowest among the evaluated models. These results show that v11n detected more objects and missed fewer objects than the other models. It also obtained the highest mAP50 and mAP50–95 values among the evaluated models.
<br>

**YOLO v26n** <br>
v26n achieved 89.96% Precision, 86.23% Recall, 91.26% mAP@50, and 77.47% mAP@50–95. Its False Alarm Rate increased to 10.04%, while its Miss Rate was 13.77%. These results show that v26n produced more false detections than the other models and had the lowest Precision among the evaluated models.

# **CONCLUSION**
The model selected for deployment in the Android application is determined using the evaluation criteria listed below:
1. mAP@50 - measures the model's ability to correctly detect and locate objects when a predicted bounding box overlaps the actual object by at least 50%
2. Recall - measures how many of the objects present in the scene are successfully detected by the model
3. mAP@50-95 - measures the model's detection and localization performance across multiple overlap thresholds ranging from 50% to 95%
4. Precision - measures how often the model is correct when it detects an object
5. False Alarm Rate - measures how often the model detects an object that is not actually present
6. Miss Rate - measures how often the model fails to detect an object that is actually present

*7. Frames Per Second (FPS) - measures how many image frames the model can process in one second*
